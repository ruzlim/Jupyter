```python
import matplotlib.pyplot as plt
import numpy as np
```

## Bar Chart

```python
x = ['Group A','Group B','Group C']
y = [10,20,30]

plt.figure(figsize=(5,3))
# plt.figure(dpi=150)
# plt.figure(figsize=(5,3), dpi=150)

plt.bar(x, y) 

plt.xticks(x)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')  
# plt.savefig('myBar.png', dpi=300)
plt.show()
```

```python
x = ['Group A','Group B','Group C','Group D']
y = [10,20,30,40]

plt.figure(figsize=(5,3))

bars = plt.bar(x, y)

# bars[0].set_hatch('/')
# bars[1].set_hatch('.')
# bars[2].set_hatch('o')
# bars[3].set_hatch('*')
patterns = ['/', '.', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# plt.xticks(x)
# plt.xlabel('X')
# plt.ylabel('Y')
# plt.title('Title')
plt.show()
```

```python
x = ['Group A','Group B','Group C','Group D']
y = [10,20,20,40]
z = [20,30,25,30]

x_axis = np.arange(len(x))

plt.figure(figsize=(5,3))

plt.bar(x_axis - 0.2, y, 0.4, label = 'Y label')
plt.bar(x_axis + 0.2, z, 0.4, label = 'Z label')

plt.xticks(x_axis, x)
plt.xlabel('Groups')
plt.ylabel('Number of Values')
plt.title('Number of Values in each group')
plt.legend()
plt.show()
```

## Sub Plot

```python
# plt.figure(figsize=(5,3))

# # Some example data to display
# x = np.linspace(0, 2 * np.pi, 400)
# y = np.sin(x ** 2)

x = ['Group A','Group B','Group C','Group D']
# y = [10,20,20,40]
y = np.asarray([10,20,20,40])
z = [10,100,30,70]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
```

```python
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
axs[2].plot(x, z)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
ax3.plot(x, z)
```

## Pie Chart

```python
y = np.array([2, 3])
mylabels = ["A", "B"]
mycolors = ["orangered", "dodgerblue"]
# myexplode = [0.2, 0]

plt.title('Title', fontdict={'fontweight': 'bold'})
plt.pie(y, labels=mylabels, colors=mycolors, autopct='%1.1f%%', wedgeprops = {"linewidth": 3, "edgecolor": "white"})
# plt.legend()
# plt.tight_layout()
plt.show()
```
